### QUESTION 1: What is hypothesis testing in statistics?

**Hypothesis Testing Overview**:  
Hypothesis testing is a statistical procedure used to evaluate if there is enough evidence in a sample of data to support a specific hypothesis about a population. This is done by comparing a sample statistic to a hypothesized population parameter, then determining whether the observed sample provides sufficient evidence to reject or fail to reject the null hypothesis.

**Steps in Hypothesis Testing**:
1. **Formulating Hypotheses**:
   - **Null Hypothesis (H₀)**: This is the hypothesis that suggests no effect, no difference, or no relationship between the variables.
   - **Alternative Hypothesis (H₁ or Ha)**: This hypothesis suggests that there is an effect, difference, or relationship.

2. **Setting the Significance Level (α)**:  
   The significance level (α) determines how much risk you're willing to accept of making a Type I error (i.e., rejecting the null hypothesis when it is true). Common choices for α are 0.05, 0.01, or 0.10.

3. **Selecting a Test Statistic**:  
   Based on the type of hypothesis and data, choose an appropriate test statistic (e.g., Z-test, T-test, Chi-square test).

4. **Calculate the P-value**:  
   The P-value is the probability of obtaining results at least as extreme as the results actually observed, given that the null hypothesis is true.

5. **Make a Decision**:
   - If the P-value ≤ α, reject the null hypothesis.
   - If the P-value > α, fail to reject the null hypothesis.

**Example**:  
Suppose a company claims their new smartphone's battery lasts more than 10 hours on average. You want to test whether this claim is true, so you perform hypothesis testing.

- **Null Hypothesis (H₀)**: The average battery life is 10 hours (μ = 10).
- **Alternative Hypothesis (H₁)**: The average battery life is greater than 10 hours (μ > 10).

After collecting sample data, you compute the test statistic and P-value, and decide whether the data supports the company's claim.

### QUESTION 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Null Hypothesis (H₀)**:  
- The null hypothesis is the default assumption that there is no effect, no difference, or no relationship between the variables in the population.
- It is the hypothesis that we aim to test and potentially reject.

**Alternative Hypothesis (H₁ or Ha)**:  
- The alternative hypothesis represents the opposite of the null hypothesis. It suggests that there is a true effect, difference, or relationship in the population.

**Key Differences**:
- **H₀**: Assumes no change or difference.
- **H₁**: Suggests there is a significant change or difference.
- Hypothesis testing is about determining whether the sample data provides enough evidence to reject H₀ in favor of H₁.

**Example**:
- **Null Hypothesis (H₀)**: The average salary in a company is $50,000.
- **Alternative Hypothesis (H₁)**: The average salary in the company is not $50,000.

You would collect data (e.g., salary data from a sample of employees) and then test whether the sample data provides sufficient evidence to reject H₀.

### QUESTION 3: What is the significance level in hypothesis testing, and why is it important?

**Significance Level (α)**:  
- The significance level (α) is a threshold value that defines how much risk of making a Type I error is acceptable in hypothesis testing. It represents the probability of rejecting the null hypothesis when it is actually true.
- Common values for α are 0.05, 0.01, or 0.10, with 0.05 being the most commonly used.

**Why It Is Important**:
- **Controls Type I Error Rate**: The significance level helps control the likelihood of incorrectly rejecting the null hypothesis (Type I error).
- **Decision Making**: The significance level is used to compare with the P-value to decide whether the null hypothesis should be rejected or not.

**Example**:
- Suppose you perform a test with α = 0.05. If you get a P-value of 0.03, this means that there is only a 3% chance of observing data as extreme as yours, assuming the null hypothesis is true. Since 0.03 < 0.05, you reject the null hypothesis, concluding that there is a significant effect.
- If you set α = 0.01 and get a P-value of 0.03, you would fail to reject the null hypothesis because 0.03 > 0.01.

### QUESTION 4: What does a P-value represent in hypothesis testing?

**P-value Overview**:  
- The P-value is the probability of observing results at least as extreme as those in your sample, assuming the null hypothesis is true. It measures the strength of the evidence against the null hypothesis.
- A smaller P-value indicates stronger evidence against the null hypothesis.

**Interpretation**:
- **P-value ≤ α**: Reject the null hypothesis. The data provides enough evidence to support the alternative hypothesis.
- **P-value > α**: Fail to reject the null hypothesis. The data does not provide enough evidence to support the alternative hypothesis.

**Example**:  
You are testing whether a new drug lowers cholesterol levels:
- **Null Hypothesis (H₀)**: The drug has no effect on cholesterol.
- **Alternative Hypothesis (H₁)**: The drug lowers cholesterol.

After conducting the test, you find a P-value of 0.03. If you set α = 0.05, the P-value is smaller than α, so you reject the null hypothesis and conclude that the drug likely lowers cholesterol.

### QUESTION 5: How do you interpret the P-value in hypothesis testing?

**Interpretation of P-value**:
- **P-value ≤ α**: The results are statistically significant. This means the observed data is highly unlikely under the assumption that the null hypothesis is true, and you reject the null hypothesis.
- **P-value > α**: The results are not statistically significant. This means the observed data is not surprising if the null hypothesis were true, and you fail to reject the null hypothesis.

**Practical Examples**:
1. **Example 1**: You are testing if a new teaching method improves student performance.
   - Null Hypothesis (H₀): The teaching method does not improve performance.
   - After testing, the P-value is 0.04, and you set α = 0.05.
   - Since 0.04 < 0.05, you reject H₀ and conclude that the teaching method improves performance.

2. **Example 2**: Testing if a coin is biased towards heads.
   - Null Hypothesis (H₀): The coin is fair (probability of heads is 0.5).
   - After flipping the coin 100 times, you find a P-value of 0.15.
   - Since 0.15 > 0.05, you fail to reject H₀ and conclude that the coin is not biased.

### QUESTION 6: What are Type 1 and Type 2 errors in hypothesis testing?

**Type 1 Error (False Positive)**:  
- Occurs when you **reject the null hypothesis** when it is actually **true**.
- **Consequences**: You mistakenly conclude that there is an effect or difference when there isn’t.
- **Probability**: This is denoted by α, the significance level.

**Type 2 Error (False Negative)**:  
- Occurs when you **fail to reject the null hypothesis** when it is actually **false**.
- **Consequences**: You mistakenly conclude that there is no effect or difference when there actually is.
- **Probability**: This is denoted by β, and the power of the test is 1 - β.

**Example**:  
Imagine testing whether a drug improves recovery time:
- **Type 1 Error**: Concluding that the drug works (rejecting H₀) when it actually doesn’t.
- **Type 2 Error**: Concluding that the drug doesn’t work (failing to reject H₀) when it actually does.

### QUESTION 7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

**One-Tailed Test**:  
- Used when the alternative hypothesis specifies a direction of the effect.
- **Example**: Testing if a new drug increases recovery time.  
  - **Null Hypothesis (H₀)**: The drug has no effect (μ = 0).  
  - **Alternative Hypothesis (H₁)**: The drug increases recovery time (μ > 0).

**Two-Tailed Test**:  
- Used when the alternative hypothesis does **not** specify a direction.  
- **Example**: Testing if a new drug has a different effect on recovery time, without assuming if it increases or decreases.
  - **Null Hypothesis (H₀)**: The drug has no effect (μ = 0).  
  - **Alternative Hypothesis (H₁)**: The drug has a different effect (μ ≠ 0).

**Key Difference**:  
- **One-Tailed**: Tests for an effect in **one direction**.
- **Two-Tailed**: Tests for an effect in **both directions**.

### QUESTION 8: What is the Z-test, and when is it used in hypothesis testing?

**Z-test Overview**:  
- The Z-test is used when the population variance is known or when the sample size is large enough (typically n > 30) to approximate the population distribution to a normal distribution.
- It is used for testing hypotheses about the **mean** of a population.

**When to Use**:
- Population standard deviation is known.
- Sample size is large (n > 30).
- Data is approximately normally distributed.

**Formula**:  
\[
Z = \frac{X̄ - μ_0}{\frac{σ}{\sqrt{n}}}
\]
Where:
- \( X̄ \) = sample mean
- \( μ_0 \) = hypothesized population mean
- \( σ \) = population standard deviation
- \( n \) = sample size

**Example**:  
You want to test if the average height of students in a school is 170 cm:
- **Null Hypothesis (H₀)**: The average height is 170 cm.
- **Alternative Hypothesis (H₁)**: The average height is not 170 cm.
Using the Z-test, you calculate the test statistic and compare it to the critical value to decide whether to reject H₀.

### QUESTION 9: How do you calculate the Z-score, and what does it represent in hypothesis testing?

**Z-score Overview**:  
- The **Z-score** measures how many standard deviations an individual data point is away from the **mean** of a distribution. It is a way of standardizing data for comparison.

**Formula**:  
\[
Z = \frac{X - μ}{σ}
\]
Where:
- \( X \) = individual data point
- \( μ \) = population mean
- \( σ \) = population standard deviation

**Interpretation**:
- A Z-score of 0 means the data point is exactly at the mean.
- A Z-score of 1 means the data point is 1 standard deviation above the mean.
- A Z-score of -1 means the data point is 1 standard deviation below the mean.

**Example**:
In a test with a mean score of 70 and a standard deviation of 10:
- A score of 80 would have a Z-score of:
\[
Z = \frac{80 - 70}{10} = 1
\]
- This means the score is 1 standard deviation above the mean.

### QUESTION 10: What is the T-distribution, and when should it be used instead of the normal distribution?

**T-distribution Overview**:  
- The **T-distribution** is similar to the normal distribution but has **fatter tails**. This means it accounts for more variability when estimating the population mean with small sample sizes.
- It is used when the sample size is **small** (n < 30) or when the population standard deviation is **unknown**.

**When to Use**:
- Small sample size (n < 30).
- Population standard deviation is unknown.
- Data is approximately normally distributed.

**Key Difference from Normal Distribution**:
- The T-distribution is more spread out and has heavier tails, making it more suitable for small sample sizes.
- As sample size increases, the T-distribution approaches the normal distribution.

**Example**:
You have a small sample of 15 people and you want to test if their average weight is 70 kg. Since the sample size is small and you don’t know the population standard deviation, you should use the **T-distribution** instead of the normal distribution.

### QUESTION 11: What is the difference between a Z-test and a T-test?

**Z-test vs. T-test**:  
The **Z-test** and the **T-test** are both statistical methods used for hypothesis testing, but they differ mainly in the assumptions about the data:

#### **Z-test**:
- Used when the **population standard deviation** is known.
- Best suited for **large sample sizes** (n > 30).
- Assumes that the data is **normally distributed**.
- The test statistic follows the **normal distribution**.

#### **T-test**:
- Used when the **population standard deviation** is **unknown** and has to be estimated from the sample.
- Best suited for **small sample sizes** (n < 30).
- The test statistic follows the **t-distribution**, which has **heavier tails** to account for increased variability in smaller samples.

**Key Difference**:  
- **Z-test**: Known population standard deviation, large sample sizes.
- **T-test**: Unknown population standard deviation, small sample sizes.

### QUESTION 12: What is the T-test, and how is it used in hypothesis testing?

**T-test Overview**:  
The **T-test** is used to determine if there is a significant difference between the means of two groups or if the sample mean significantly differs from a known population mean. It is particularly useful when dealing with small sample sizes.

#### Types of T-tests:
1. **One-Sample T-test**: Tests if the sample mean differs from a known population mean.
   - Example: Testing if the average salary of employees in a company is 50,000.

2. **Independent Two-Sample T-test**: Compares the means of two independent groups.
   - Example: Comparing the average test scores of two different classes.

3. **Paired Sample T-test**: Compares the means of two related groups.
   - Example: Testing if a group’s blood pressure changes before and after medication.

**Formula for One-Sample T-test**:
\[
t = \frac{X̄ - μ_0}{\frac{s}{\sqrt{n}}}
\]
Where:
- \( X̄ \) = sample mean
- \( μ_0 \) = hypothesized population mean
- \( s \) = sample standard deviation
- \( n \) = sample size

### QUESTION 13: What is the relationship between the Z-test and T-test in hypothesis testing?

**Relationship between Z-test and T-test**:  
- Both tests are used for **comparing sample means** to a population mean or between two sample means.
- The main difference lies in the assumptions:
  - **Z-test** is used when the **population standard deviation** is known or for **large sample sizes** (n > 30).
  - **T-test** is used when the **population standard deviation** is unknown and when the sample size is small (n < 30).
  
As the sample size increases, the **T-distribution** approaches the **normal distribution**, and the **T-test** becomes almost identical to the **Z-test** for large sample sizes.

### QUESTION 14: What is a confidence interval, and how is it used to interpret statistical results?

**Confidence Interval (CI) Overview**:  
A **confidence interval** is a range of values, derived from sample data, that is likely to contain the population parameter (e.g., mean) with a certain degree of confidence.

#### Key Concepts:
- The **confidence level** (usually 95% or 99%) indicates the probability that the interval contains the true population parameter.
- A **wider confidence interval** means less precision in the estimate, while a **narrower confidence interval** indicates more precision.
  
**Formula for CI for a Mean** (with known population standard deviation):
\[
CI = X̄ \pm Z \times \frac{σ}{\sqrt{n}}
\]
Where:
- \( X̄ \) = sample mean
- \( Z \) = Z-value corresponding to the desired confidence level
- \( σ \) = population standard deviation
- \( n \) = sample size

**Example**:  
If you calculate a 95% CI for the average height of a population to be (170 cm, 175 cm), this means you are 95% confident that the true population mean falls within this interval.

### QUESTION 15: What is the margin of error, and how does it affect the confidence interval?

**Margin of Error (MoE) Overview**:  
The **margin of error** is the amount of error allowed in the estimation of a population parameter. It indicates how much the sample estimate might differ from the true population value.

#### Relationship with Confidence Interval:
- The **margin of error** is half the width of the confidence interval.
- A larger margin of error means a **wider** confidence interval, indicating **less precision**.
- A smaller margin of error means a **narrower** confidence interval, indicating **more precision**.

**Formula for Margin of Error**:
\[
MoE = Z \times \frac{σ}{\sqrt{n}}
\]
Where:
- \( Z \) = Z-value corresponding to the desired confidence level.
- \( σ \) = population standard deviation.
- \( n \) = sample size.

**Example**:  
If a sample mean is 50 and the margin of error is 5, the confidence interval would be (45, 55), meaning the true population mean is expected to fall between 45 and 55 with a certain level of confidence.

### QUESTION 16: How is Bayes' Theorem used in statistics, and what is its significance?

**Bayes' Theorem Overview**:  
Bayes' Theorem describes how to update the probability of a hypothesis based on new evidence. It is particularly useful in **statistical inference** and **decision-making** where you have prior knowledge (prior probability) and you update it with new data (likelihood).

#### Formula for Bayes' Theorem:
\[
P(H|E) = \frac{P(E|H) \times P(H)}{P(E)}
\]
Where:
- \( P(H|E) \) = **posterior probability** (probability of the hypothesis \( H \) given evidence \( E \)).
- \( P(E|H) \) = **likelihood** (probability of the evidence given the hypothesis).
- \( P(H) \) = **prior probability** (initial belief about the hypothesis).
- \( P(E) \) = **marginal likelihood** (probability of the evidence).

#### Significance:
- **Bayesian Inference**: Bayes’ theorem allows you to continuously update your beliefs about a hypothesis as more data becomes available.
- **Machine Learning**: It is widely used in classification algorithms (e.g., Naive Bayes classifiers), where you calculate the probability of different classes given the input data.

**Example**:
If you have a prior belief that 1% of people have a rare disease, and a diagnostic test has a 95% accuracy rate, Bayes' theorem can help you update the probability of a person having the disease given a positive test result.

### QUESTION 17: What is the Chi-square distribution, and when is it used?

**Chi-square Distribution Overview**:  
The **Chi-square distribution** is a probability distribution that arises when summing the squares of independent standard normal variables. It is used in hypothesis testing, particularly in tests of independence and goodness of fit.

#### Key Properties:
- The **Chi-square distribution** is **positively skewed** and its shape depends on the degrees of freedom (df).
- It only takes **positive values**.
- As the degrees of freedom increase, the distribution becomes more symmetrical and approaches a normal distribution.

#### When to Use:
- **Chi-square Goodness of Fit Test**: To determine if a sample data fits a hypothesized distribution (e.g., testing if a die is fair).
- **Chi-square Test of Independence**: To determine if two categorical variables are independent (e.g., testing if gender is independent of voting preference).

---

### QUESTION 18: What is the Chi-square goodness of fit test, and how is it applied?

**Chi-square Goodness of Fit Test Overview**:  
The **Chi-square goodness of fit test** is used to determine if the observed frequencies in a categorical dataset match the expected frequencies under a specific hypothesis.

#### Steps in the Test:
1. **State the Hypotheses**:
   - **Null Hypothesis (H₀)**: The observed frequencies match the expected frequencies.
   - **Alternative Hypothesis (H₁)**: The observed frequencies do not match the expected frequencies.

2. **Calculate the Test Statistic**:
   \[
   χ² = \sum \frac{(O_i - E_i)^2}{E_i}
   \]
   Where:
   - \( O_i \) = observed frequency in the i-th category.
   - \( E_i \) = expected frequency in the i-th category.

3. **Determine the Degrees of Freedom**:
   \[
   df = k - 1
   \]
   Where \( k \) is the number of categories.

4. **Compare the Test Statistic to the Critical Value** from the Chi-square distribution table at the chosen significance level.

5. **Make a Decision**: If the test statistic is greater than the critical value, reject the null hypothesis.

**Example**:
You want to test if a die is fair. The expected frequency for each side of the die (for 60 rolls) is 10. If the observed frequencies are significantly different, the Chi-square test will help determine if the die is biased.

### QUESTION 19: What is the F-distribution, and when is it used in hypothesis testing?

**F-distribution Overview**:  
The **F-distribution** is a probability distribution used in hypothesis testing, primarily for comparing variances. It is the ratio of two independent Chi-square variables, each divided by their respective degrees of freedom.

#### Key Properties:
- The **F-distribution** is **positively skewed**.
- It is used to test if the variances of two or more groups are equal.
- The shape of the F-distribution depends on the degrees of freedom of the numerator and denominator.

#### When to Use:
- **ANOVA (Analysis of Variance)**: The F-distribution is used to test if the means of several groups are equal by comparing variances.
- **Comparing Variances**: The F-test compares the variances of two groups to determine if they are significantly different.

### QUESTION 20: What is an ANOVA test, and what are its assumptions?

**ANOVA (Analysis of Variance)** Overview:  
**ANOVA** is a statistical test used to compare the means of three or more groups to determine if at least one group mean is different from the others.

#### Types of ANOVA:
1. **One-Way ANOVA**: Compares the means of three or more groups based on one independent variable.
2. **Two-Way ANOVA**: Compares the means of multiple groups based on two independent variables.

#### Assumptions of ANOVA:
1. **Independence**: Observations in each group are independent.
2. **Normality**: The data in each group are approximately normally distributed.
3. **Homogeneity of Variances**: The variance within each group is approximately equal.

#### Steps in ANOVA:
1. **State the Hypotheses**:
   - **Null Hypothesis (H₀)**: All group means are equal.
   - **Alternative Hypothesis (H₁)**: At least one group mean is different.

2. **Calculate the F-statistic**:
   \[
   F = \frac{\text{Between-group variance}}{\text{Within-group variance}}
   \]

3. **Compare the F-statistic** with the critical value from the F-distribution table to make a decision.

**Example**:  
Suppose you want to compare the test scores of students from three different teaching methods. ANOVA will help determine if there is a statistically significant difference in the mean scores.

### QUESTION 21: What are the different types of ANOVA tests?

**Types of ANOVA Tests**:  
ANOVA (Analysis of Variance) tests are used to compare the means of multiple groups. There are several types of ANOVA, each serving a different purpose depending on the complexity of the data and the number of factors involved.

#### 1. **One-Way ANOVA**:
- **Purpose**: Compares the means of three or more independent groups based on one independent variable.
- **Example**: Comparing the average scores of three different teaching methods (e.g., Method A, Method B, Method C) to see if they affect students’ scores differently.

#### 2. **Two-Way ANOVA**:
- **Purpose**: Compares the means of multiple groups based on two independent variables (factors).
- **Example**: Investigating how both teaching method (Method A, Method B, Method C) and study time (1 hour, 2 hours, 3 hours) affect students’ scores. It can also assess if there is an **interaction effect** between the two factors.

#### 3. **Repeated Measures ANOVA**:
- **Purpose**: Compares means for the same group at different times or under different conditions. It is used when the same subjects are used in each condition.
- **Example**: Measuring the effectiveness of a drug by comparing the blood pressure of patients before treatment, during treatment, and after treatment.

#### 4. **Multivariate Analysis of Variance (MANOVA)**:
- **Purpose**: Similar to ANOVA but allows for the analysis of multiple dependent variables simultaneously.
- **Example**: Analyzing the effect of different teaching methods on both students’ scores and their overall satisfaction with the course.

### QUESTION 22: What is the F-test, and how does it relate to hypothesis testing?

**F-test Overview**:  
The **F-test** is used to compare two variances or to test if the variances of two populations are equal. It is commonly used in **ANOVA** tests, as the F-test assesses the ratio of variances between groups.

#### When to Use:
- **Comparing Variances**: The F-test is used when we want to test whether two populations have the same variance.
- **In ANOVA**: The F-test in ANOVA helps determine if the means of different groups are significantly different by comparing the variability between the groups to the variability within the groups.

#### Formula for F-test:
\[
F = \frac{\text{Variance between groups}}{\text{Variance within groups}}
\]

- **Between-group variance**: Measures how the means of the groups differ from the overall mean.
- **Within-group variance**: Measures how much variability exists within each group.

#### Interpretation:
- A **high F-statistic** suggests that the means between the groups are significantly different, as the between-group variance is large compared to the within-group variance.
- A **low F-statistic** indicates that the group means are similar, and the variability within groups is greater than between groups.

**Example**:  
In an ANOVA test comparing the effect of three teaching methods on student performance, the F-test will tell us whether the variability between the methods is significantly larger than the variability within each method.

```python
# QUESTION 5: Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def type_errors(population_mean, population_std, sample_size, alternative_mean, alpha=0.05):
    # Type I Error (False Positive) is the area under the curve where we reject a true null hypothesis
    z_critical = norm.ppf(1 - alpha)

    # Type II Error (False Negative) is the area under the curve where we fail to reject a false null hypothesis
    z_beta = (population_mean - alternative_mean) / (population_std / np.sqrt(sample_size))

    # Plotting
    x = np.linspace(population_mean - 5, population_mean + 5, 1000)
    y = norm.pdf(x, population_mean, population_std)

    plt.plot(x, y)
    plt.fill_between(x, 0, y, where=(x > population_mean + z_critical), color='red', alpha=0.3, label="Type I Error")
    plt.fill_between(x, 0, y, where=(x < population_mean - z_beta), color='blue', alpha=0.3, label="Type II Error")

    plt.title('Type I and Type II Errors')
    plt.legend()
    plt.show()

type_errors(0, 1, 30, 2)
```

```python
# QUESTION 10: Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats

def margin_of_error(data, confidence=0.95):
    mean = np.mean(data)
    std_error = stats.sem(data)
    critical_value = stats.t.ppf((1 + confidence) / 2, len(data) - 1)
    margin_of_error = critical_value * std_error
    return margin_of_error

data = np.random.normal(50, 10, 100)
me = margin_of_error(data)
print(f'Margin of Error: {me}')
```

```python
# QUESTION 11: Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
import numpy as np

def bayesian_inference(prior, likelihood, evidence):
    # Bayes' Theorem: P(H|E) = (P(E|H) * P(H)) / P(E)
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.5  # Initial belief (prior probability)
likelihood = 0.8  # Probability of evidence given the hypothesis
evidence = 0.7  # Total probability of evidence

posterior = bayesian_inference(prior, likelihood, evidence)
print(f'Posterior Probability: {posterior}')
```

```python
# QUESTION 13: Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data
import numpy as np

def expected_frequencies(observed_data):
    total = np.sum(observed_data)
    row_totals = np.sum(observed_data, axis=1)
    col_totals = np.sum(observed_data, axis=0)
    expected = np.outer(row_totals, col_totals) / total
    return expected

observed_data = np.array([[10, 20, 30],
                          [20, 15, 35]])

expected = expected_frequencies(observed_data)
print(f'Expected Frequencies: \n{expected}')
```

```python
# QUESTION 16: Implement an F-test using Python to compare the variances of two random samples
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2):
    # Calculate the ratio of variances
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    f_stat = var1 / var2
    df1 = len(sample1) - 1
    df2 = len(sample2) - 1

    # Compute the p-value for the F-distribution
    p_value = 1 - f.cdf(f_stat, df1, df2)
    return f_stat, p_value

sample1 = np.random.normal(50, 10, 100)
sample2 = np.random.normal(55, 12, 100)

f_stat, p_value = f_test(sample1, sample2)
print(f'F-statistic: {f_stat}, P-value: {p_value}')
```

```python
# QUESTION 19: Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
import numpy as np
from scipy.stats import normaltest, levene

def check_anova_assumptions(*groups):
    # Check normality assumption (using D'Agostino and Pearson's test)
    normality_results = [normaltest(group) for group in groups]

    # Check equal variance assumption (using Levene's test)
    equal_variance = levene(*groups)

    return normality_results, equal_variance

# Simulate data for three groups
group1 = np.random.normal(50, 10, 30)
group2 = np.random.normal(55, 10, 30)
group3 = np.random.normal(60, 10, 30)

normality_results, equal_variance = check_anova_assumptions(group1, group2, group3)
print(f'Normality Test Results: {normality_results}')
print(f'Equal Variance Test (Levene\'s Test) Result: {equal_variance}')
```
